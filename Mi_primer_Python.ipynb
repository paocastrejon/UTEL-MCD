La metodología de investigación es una disciplina que guía la investigación, no un simple protocolo. Es estable y estandarizada, lo que permite compartir el conocimiento globalmente, aunque ha evolucionado y ofrece diversos métodos y técnicas. Existen diferentes formas de clasificar la investigación, enseguida se presentan los diferentes esquemas de tipificación de acuerdo con Behar (2010) y Quezada, Apolo y Delgado (2018):

**Con base en el propósito o finalidad**

La investigación se divide principalmente en dos tipos:

Investigación básica (o pura): Se enfoca en crear nuevas teorías y conocimientos generales. No busca una aplicación inmediata, sino entender los "porqués" de las cosas. ¡Piensa en ella como la base del saber!

Investigación aplicada: Su objetivo es resolver problemas reales usando el conocimiento ya existente (a menudo, el que proviene de la investigación básica). ¡Es la que pone la teoría en práctica!

Muchas investigaciones son mixtas, combinando aspectos teóricos y prácticos.

**Con base en los medios utilizados para la obtención de datos**

* Documental: Consiste en recopilar y analizar información de documentos ya existentes, como libros, artículos, archivos o sitios web. ¡Piensa en ello como investigar en la biblioteca o internet!

* De Campo: Se obtiene la información directamente del lugar donde ocurren los hechos. Esto se logra con herramientas como la observación, entrevistas o encuestas. ¡Es salir a "ver" lo que pasa!

* Experimental: Aquí, los investigadores recrean un fenómeno en condiciones controladas para obtener datos precisos y poder repetir el experimento. ¡Es como montar un laboratorio para entender algo!

**Con base en el nivel o profundidad del conocimiento**

los tipos de investigación según su propósito:

* Exploratoria: Es el primer acercamiento a un tema. Busca ver qué se sabe del problema para planear estudios más profundos. ¡Es como un "reconocimiento" inicial!

* Descriptiva: Su meta es detallar y describir a fondo un fenómeno, comunidad o problema. Te da una imagen clara de la situación. ¡Piénsala como una "fotografía" detallada!

* Explicativa: Se centra en descubrir las causas de un problema o fenómeno. Explica por qué ocurren las cosas y cómo se relacionan. ¡Es como responder el "por qué" de la situación!

**Con base en el paradigma y tratamiento de los datos utilizados**

los tipos de investigación según su enfoque:

* Cualitativa: Se centra en observar y entender aspectos que no se pueden medir con números, como comportamientos, opiniones o experiencias. ¡Piensa en explorar "calidades"!

* Cuantitativa: Se enfoca en medir y usar datos numéricos para encontrar patrones y explicaciones generales. ¡Aquí buscas "cantidades" y estadísticas!

**Con base en el grado de manipulación de las variables**

Experimental: Busca causa y efecto controlando todo rigurosamente. ¡Es como un experimento de laboratorio con reglas estrictas!

Cuasi-experimental: Se parece a la experimental, pero no puede controlar todas las variables a la perfección. Los resultados son un poco menos precisos. ¡Es un experimento "casi" perfecto!

No experimental: Solo observa el fenómeno tal como es, sin manipular nada. ¡Aquí solo miras, no intervienes!
